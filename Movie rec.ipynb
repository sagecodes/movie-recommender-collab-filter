### Collaborative Filtering

```python
ratings_mean_count_df
```

```python
movie_rating_df
```

```python
userid_movietitle_matrix = movie_rating_df.pivot_table(index = 'user_id', columns = 'title', values = 'rating')
```

```python
userid_movietitle_matrix
```

```python
star_wars = userid_movietitle_matrix['Star Wars (1977)']
```

```python
star_wars
```

```python
titanic = userid_movietitle_matrix['Titanic (1997)']
```

```python
titanic
```

```python
titanic_correlations = pd.DataFrame(userid_movietitle_matrix.corrwith(titanic), columns = ['Correlation'])
```

```python
titanic_correlations = titanic_correlations.join(ratings_mean_count_df['count'])
```

```python
titanic_correlations
```

```python
titanic_correlations.dropna(inplace = True)
titanic_correlations
```

```python
titanic_correlations.sort_values('Correlation', ascending=False)
```

```python
titanic_correlations[titanic_correlations['count']>80].sort_values('Correlation', ascending = False)
```

```python
starwars_correlations = pd.DataFrame(userid_movietitle_matrix.corrwith(star_wars), columns = ['Correlation'])
```

```python
starwars_correlations.dropna(inplace = True)
```

```python
starwars_correlations = starwars_correlations.join(ratings_mean_count_df['count'])
```

```python
starwars_correlations
```

```python
starwars_correlations[starwars_correlations['count']>80].sort_values('Correlation', ascending = False)
```

### Filter for whole dataset

```python
userid_movietitle_matrix
```

```python
movie_correlations = userid_movietitle_matrix.corr(method='pearson', min_periods = 80)
```

```python
movie_correlations
```

```python
my_ratings = pd.read_csv('../datasets/movies/My_ratings.csv')
```

```python
my_ratings
```

```python
my_ratings['Movie Name'][0]
```

```python
similar_movies_list = pd.Series()

for i in range(0,2):
    similar_movie = movie_correlations[my_ratings['Movie Name'][i]].dropna()
    similar_movie = similar_movie.map(lambda x: x* my_ratings['Ratings'][i])
    similar_movies_list = similar_movies_list.append(similar_movie)
```

```python
similar_movies_list.sort_values(inplace=True, ascending=False)
print(similar_movies_list.head(10))
```
